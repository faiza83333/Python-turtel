### 4. Dessiner un Carré en utilisant une boucle

Les premiers ordinateurs ont été créés pour effectuer des calculs répétitifs et soulager le travail des scientifiques ou des ingénieurs.  
**Les boucles permettent de répéter une même instruction plusieurs fois.**

<blockquote style="color:blue;">
  <strong>Chaque répétition s'appelle une itération.</strong>
</blockquote>

#### La boucle `for`

À chaque fois qu'il est nécessaire de répéter **une action** un certain nombre de fois, on utilise une boucle

Par exemple, pour dessiner un carré,**l'action c'est avancer et tourner**

Pour pour faciliter la tâche, on utilise la boucle pour répéter 4 fois **l'itération**. 4 fois l'action, avancer et tourner.

#### Comprendre la notion d'itération

Pour mieux comprendre la notion d'itération, nous allons changer la couleur à chaque fois qu'une itération est terminée. Par exemple, 

pour dessiner un carré, nous avons besoin de répéter **4 fois l'action d'avancer et de tourner**. Après chaque itération (c'est-à-dire après chaque action d'avancer et tourner), nous changeons la couleur du tracé. Cela permet de visualiser les différentes étapes du dessin et de comprendre comment une boucle fonctionne.



Prenez l'habitude, chaque fois que vous voulez utiliser une boucle, de dessiner un tableau. Ce tableau doit comporter une 
**colonne pour le nombre d'itérations (le nombre de répétitions des actions)** ainsi qu'une **colonne pour le résultat attendu**. 

Dans cet exercice, nous l'utilisons pour dessiné les côté, mais dans les prochains cours, cela sera utile pour d'autres types de résultats.

Ce tableau sera essentiel, surtout au début de votre pratique en tant que programmeur



| Itération | côté dessiné
|-----------|------------------|
| 0           #1er côté        |
| 1         | #2er côté        |
| 2         | #3er côté        |
| 3         | #4er côté             

#### La boucle `While`

while peut se traduire par **tant que**. Les instructions dans le bloc qui suit while se répéteront tant que la condition donnée à while est vraie (on ne sait pas forcément combien il y aura d'itérations,


Voyons comment dessiner un carré simple en utilisant une boucle pour éviter la répétition de code.

```python
from turtle import *

color("red") # Définir la couleur du crayon
width(5) # Définir l'épaisseur du crayon
# Boucle pour dessiner un carré
for i in range(4):
    forward(50)  # Avancer de 'taille' pixels
    right(90)  # Tourner à droite de 90 degrés
    font = ("Arial", 24, "normal")  # (font family, font size, font style)
    write(i,font=font)
done()
```

```python
from turtle import *

color("red") # Définir la couleur du crayon
width(5) # Définir l'épaisseur du crayon
# Boucle pour dessiner un carré
i=0
while i<5 :
    forward(50)  # Avancer de 'taille' pixels
    right(90)  # Tourner à droite de 90 degrés
    font = ("Arial", 24, "normal")  # (font family, font size, font style)
    write(i,font=font)
    i=i+1
done()
```

### 5.Dessin un Carré en utilisant une foction

Si vous voulez dessiner plusieurs carrés, pour éviter de répéter plusieurs fois le même code, il est possible de minimiser le nombre d'instructions en utilisant une fonction. 

Définir une fonction permet de l'appeler à chaque fois que nécessaire, ce qui évite les répétitions et rend le code plus efficace

```python
from turtle import *
# Fonction pour dessiner un carré
def dessiner_carre(taille):
    color("red")  # Définir la couleur du crayon
    width(5)  # Définir l'épaisseur du crayon
    # Boucle pour dessiner un carré
    for i in range(4):
        forward(taille)  # Avancer de 'taille' pixels
        right(90)  # Tourner à droite de 90 degrés
        
dessiner_carre(50)
dessiner_carre(100)
dessiner_carre(150)
done()
```

```python
from turtle import *

# Fonction pour dessiner un carré
def dessiner_carre(taille, couleur, epaisseur):
    color(couleur)  # Définir la couleur du crayon
    width(epaisseur)  # Définir l'épaisseur du crayon
    
    # Boucle pour dessiner un carré
    for i in range(4):
        forward(taille)  # Avancer de 'taille' pixels
        right(90)  # Tourner à droite de 90 degrés

dessiner_carre(50,"green",5)
dessiner_carre(100,"red",10)
dessiner_carre(150,"blue",3)
done()
```

 ### 6.Déplacer le crayon

Si vous remarquez, avec Turtle, on dessine de manière continue. Pour organiser le code et séparer les parties, comme par exemple pour dessiner plusieurs carrés, il est important de comprendre la notion de canevas. Avec Turtle, cela permet de structurer et d'optimiser le dessin


On peut déplacer le crayon avec l'instruction `goto(x, y)`, où `x` et `y` représentent les coordonnées du point de destination.
![Repère pour les coordonnées](https://zestedesavoir.com/media/galleries/2660/8b076e47-f928-47a2-ba38-810e60aa0d25.png)

Cependant, en se déplaçant, le crayon trace automatiquement un trait, ce qui ne diffère pas beaucoup de la fonction `forward` que nous avons vue précédemment.

Toutefois, Turtle offre la possibilité de :
- lever le crayon avec `up()`
- baisser le crayon avec `down()`

En combinant ces commandes avec `goto()`, il devient possible de réaliser plusieurs dessins distincts côte à côte sans laisser de traits indésirables entre eux."


```python
from turtle import *

# Fonction pour dessiner un carré
def dessiner_carre(taille, couleur, epaisseur):
    color(couleur)  # Définir la couleur du crayon
    width(epaisseur)  # Définir l'épaisseur du crayon
    
    # Boucle pour dessiner un carré
    for i in range(4):
        forward(taille)  # Avancer de 'taille' pixels
        right(90)  # Tourner à droite de 90 degrés

dessiner_carre(50,"green",5)
up()
goto(100,130)
down()
dessiner_carre(100,"red",10)
up()
goto(-100,100)
down()
dessiner_carre(50,"blue",3)
done()
```

### 4. Les conditions

Pourquoi des conditions ?

Parce qu’un programme informatique, une application donne des résultats différents en fonction.
exemple 1:

Itération 0 (i == 0) : Le premier côté du carré est dessiné en rouge avec color("red").
Itération 1 (i == 1) : Le deuxième côté est dessiné en vert avec color("green").
Itération 2 (i == 2) : Le troisième côté est dessiné en bleu avec color("blue").
Itération 3 (i == 3) : Le dernier côté est aussi dessiné en bleu avec color("blue"), 

exemple 2:

>Si utilisateur appuie sur la  touche r du clavier 
    dessiné un carré rouge 
Sinon
    dessiné un carré vert
 
 
 En python, ça s’écrit avec les mots-clefs **if** et **else **(et  **elif** ) :

```python
#exemple 01
from turtle import *

def carre_couleur():
    
    for i in range(4):
        if i==0 :
            color("red")
            forward(100)
            left(90)
        elif i==1:
            color("green")
            forward(100)
            left(90)
        elif i==2:
            color("blue")
            forward(100)
            left(90)
        else:
            color("blue")
            forward(100)
            left(90)
carre_couleur()
done()
```

```python
import turtle
from random import randint

COULEURS = ['black', 'grey', 'brown', 'orange', 'pink', 'purple',
            'red', 'blue', 'yellow', 'green']


# Fonction pour dessiner un carré
def dessiner_carre(taille, couleur, epaisseur):
    color(couleur)  # Définir la couleur du crayon
    width(epaisseur)  # Définir l'épaisseur du crayon

    # Boucle pour dessiner un carré
    for i in range(4):
        forward(taille)  # Avancer de 'taille' pixels
        right(90)  # Tourner à droite de 90 degrés


# Variable pour déplacer la tortue
X = 250

for i in range(len(COULEURS)):
    # On choisit une couleur aléatoirement
    index_choisi = randint(0, len(COULEURS) - 1)
    couleur_choisie = COULEURS[index_choisi]  # Récupérer la couleur à partir de l'index
    

    # On supprime la couleur choisie pour éviter de la rechoisir
    del COULEURS[index_choisi]

    # On se déplace pour dessiner le prochain carré

    up()  # Lever le crayon pour ne pas tracer pendant le déplacement
    goto(X, 0)  # Aller à la nouvelle position
    down()  # Rebaisser le crayon pour dessiner
    dessiner_carre(50, couleur_choisie, 5)  # Dessiner un carré de taille 50 avec l'épaisseur 5
    X +=- 60  # Ajuster le décalage entre les carrés 

# Terminer l'exécution de Turtle
done()
```
